```python
img_folder= "C:\\Users\\inn\\Desktop\\Geo_Positioning_Convnet\\Data"
IMG_WIDTH=32
IMG_HEIGHT=32

def create_dataset_PIL(img_folder):
    
    img_data_array=[]
    class_name=[]
    for file in os.listdir(img_folder):
        #for file in os.listdir(os.path.join(img_folder, dir1)):
       
        image_path= os.path.join(img_folder,  file)
        image= np.array(Image.open(image_path))
        image= np.resize(image,(IMG_HEIGHT,IMG_WIDTH,3))
        image = image.astype('float32')
        image /= 255  
        img_data_array.append(image)
        class_name.append(file)
    return img_data_array , class_name
PIL_img_data, class_name=create_dataset_PIL(img_folder)
```

```python
target_dict={k: v for v, k in enumerate(np.unique(class_name))}
target_val=  [target_dict[class_name[i]] for i in range(len(class_name))]
```

```python
model=tf.keras.Sequential()
model.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=(32, 32, 3)))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(128, (3, 3), activation='relu'))
model.add(layers.Flatten())
model.add(layers.Dense(6489, activation='relu'))
model.add(layers.Dense(4326, activation='relu'))
model.add(layers.Dense(2163, activation='relu'))


model.add(layers.Dense(721))
        
```

```python
model.compile(optimizer='rmsprop', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
history = model.fit(x=np.array(PIL_img_data, np.float32), y=(np.array(list1)), epochs=5)
```

```python
x_test=[np.array(PIL_img_data, np.float32)]
y_test=[np.array(list1)]
predictions=model.predict(x_test) 
scores = model.evaluate(x_test , y_test)
```
