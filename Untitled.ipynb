# 5 - importing libraries

```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# 6 - Reading training dataset

```python

data= pd.read_csv('train_data.csv')
data
```

# 7 - Dropping ID & HealthServiceArea Columns.

```python

data= data.drop('ID', axis=1 ).drop('HealthServiceArea', axis=1)
data
```

#  8 - Classifying the lengthofStay by Assigning 0 and 1 to the Values < 4 and >=4 Respectively.


```python

LengthOfStay = []
for i in data['LengthOfStay']:
    if i < 4:
        LengthOfStay.append(0)
    else:
        LengthOfStay.append(1)
```

```python
len(LengthOfStay), len(data['LengthOfStay'])
```

```python
data['LengthOfStay']= LengthOfStay
```

```python
data
```

#  9 - Splitting Data into X and Y.

```python

x_train=data.drop('LengthOfStay', axis=1)
y_train=data['LengthOfStay'] 
```

# 10 - Importing model.

```python

from sklearn.ensemble import RandomForestClassifier
clf=RandomForestClassifier()
```

# 11 - Converting the Non-integer Columns into Integer Columns.


```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

features_gender = ["Gender"]
features_Race = ['Race']
features_TypeOfAdmission = ['TypeOfAdmission']
features_PaymentTypology=['PaymentTypology']
features_EmergencyDepartmentIndicator= ['EmergencyDepartmentIndicator']

en_code = OneHotEncoder()

transformer = ColumnTransformer([("gender_tranform", en_code, features_gender),
                                 ("race_tranform", en_code, features_Race),
                                 ("TypeOfAdmission_tranform", en_code, features_TypeOfAdmission),
                                 ("PaymentTypology_tranform", en_code, features_PaymentTypology),
                                 ("EmergencyDepartmentIndicator_tranform", en_code, features_EmergencyDepartmentIndicator)
                                
                                ], remainder='passthrough')


transformed_x=transformer.fit_transform(x_train)
pd.DataFrame(transformed_x)

```

# 12 - Fitting the Model.

```python

clf.fit(transformed_x, y_train)
```

# 13 - Finding Accuracy

```python

clf.score(transformed_x, y_train)
```

# 14 - Importing GaussianNB column and Finding Accuracy


```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb.fit(transformed_x, y_train)
gnb.score(transformed_x, y_train)
```

As per the accuracy score RandomForestClassofier is far better than GaussianNB.
